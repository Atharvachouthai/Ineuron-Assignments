### 1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
def myreduce(function,l):
    listt = l
    new_list = listt[0]
    for i in range(1,len(l)):
        new_list = function(new_list,i)
    return new_list
    
def add(x,y):
    return x+y
   
def multiply(x,y):
    return x*y
  
l = [10,15,25,40,10]
print("The result of custom made myreduce function for addition on list is  : " +str(myreduce(add,l)))
print("The result of custom made myreduce fuction for multiplying in list is : " +str(myreduce(multiply,l)))



```

### 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
def myfilter(func,li):
    new_list = []
    for i in li:
        if func(i):
            new_list.append(i)
    return new_list
def ispositive(x):
    if (x <= 0):
        return False
    else:
        return True
```

```python
x = [1,-2,3,40,-71,33,-7]
print("The result of custom made myfilter fuction is " +str(myfilter(ispositive,x)))
```

```python
list_1 = [i*n for i in 'xyz' for n in range(1,5)]
print("The list comprehension for XYZ is : "+"  "+str(list_1))
```

```python
list_2 = [i*n for n in range(1,5) for i in 'xyz']
print("The list comprehension for XYZ is : "+"  "+str(list_2))
```

```python
list_3 = [[i+n] for i in range(2,5) for n in range(0,3)]
print("The list comprehension for 2,3,4 is : "+"  "+str(list_3))
```

```python
list_4 = [[i+n for n in range(2,6)] for i in range(0,3)]
print("The list comprehension for 2,3,4,5 is : "+"  "+str(list_4))
```

```python
list_5 = [(i,n) for n in range(1,4) for i in range(1,4)]
print("The list comprehension for 1,2,3 is : "+"  "+str(list_5))
```
