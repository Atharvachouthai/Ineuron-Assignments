### 1.1 Write a Python Program(with class concepts) to find the area of the triangle using the below
formula:
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5

Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.

```python
class sides:
    def __init__(self,s1,s2,s3):
        self.s1 = float(s1)
        self.s2 = float(s2)
        self.s3 = float(s3)        
s1 = int(input("side 1 = "))
s2 = int(input("side 2 = "))
s3 = int(input("side 3 = "))
        
class triangle(sides):
    def __init__(self,s1,s2,s3):
        super().__init__(s1,s2,s3)
        
    def get_area(self):
        s = (s1 + s2 + s3) / 2
        return (s*(s-s1)*(s-s2)*(s-s3)) ** 0.5
    
t = triangle(s1,s2,s3)
print("area : {}".format(t.get_area()))
```

### 1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.

```python
def filter_long_words(l,a):
    words = []
    for i in l:
        if(len(i)>=a):
            words.append(i)
    return words

n = input("Enter Words : ")
n_split = n.split(",")
min_len_count = input("Enter Min Length : ")
long_words = filter_long_words(n_split, int(min_len_count))

print("Word(s) with atleast"+" "+ str(min_len_count)+" " + "length : ",long_words)
```

### 2.1 Write a Python program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words.

Hint: If a list [ ab,cde,erty] is passed on to the python function output should come as [2,3,4]
Here 2,3 and 4 are the lengths of the words in the list.

```python
def mapfunc_to_count_lenofwords(l):
    m = list(map(len,l))
    return m
listt = list(input("Enter the words to populate list : ").split(","))
l = [i.strip() for i in listt]
listt = mapfunc_to_count_lenofwords(l)
print("Length of each word in list is :  ",listt)
```

### 2.2 Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.

```python
def vowel_checker(character):
    if(len(character)==1):
        vowel_list=['a','e','i','o','u']
        if(character.lower() in vowel_list):
            return True
        else:
            return False
    else:
        print("Please, enter 1 character at a time! ")

input_word = input("Enter value : ")
output_result = vowel_checker(input_word)
print("output value : ",output_result)
        
  
```
